```python
class semler_toolbox:
    def merge_delta_tables_to_core(core_target_delta_table, core_entity, merge_final_table):
        target_table_dt = DeltaTable.forName(spark, core_target_delta_table)
        target_table_df = spark.table(merge_final_table)

        target_table_dt.alias("current").merge(
            target_table_df.alias("changed"),
            f"current.{core_entity}_key = changed.{core_entity}_key"
        ).whenNotMatchedInsertAll() \
         .execute()
```
